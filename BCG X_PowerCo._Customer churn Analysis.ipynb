## Predictive Modelling :

```python
# Importing important Libraries :

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report 
```

```python
# Selecting Relevant Features :

features = [
    "cons_12m","cons_last_month","forecast_cons_12m","forecast_price_energy_peak","price_peak_fix",
           "price_mid_peak_fix","num_years_antig","margin_net_pow_ele"
]


# Defining X( Features) & y(target) :

X = merged_df[features]
y = merged_df["churn"]

# Standardising numerical features :

scale = StandardScaler()
X_scaled = scale.fit_transform(X)

# Splitting dstaset into Training (80%) testing (20%) data :


X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size = 0.2, random_state = 42)

print("Data Preprocessing process Completed")


```

## Training Logistic Regression model :

```python
# Initializing and Training the Logistic regression Model :

lrm = LogisticRegression()
lrm.fit(X_train, y_train)
print("Model Training Completed")
```

## Evaluating Model Performance :

```python
# Making Prediction :
y_pred = lrm.predict(X_test)

# Finding accuracy Score :
accuracy = accuracy_score(y_test,y_pred)
print(f"Model Accuracy Score :{accuracy:.2%}")

# Confusion Matrix
Conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", Conf_matrix)

print("Classification Report :\n", classification_report(y_test, y_pred))
```

# Feature Importance : 

```python
# Extrcting feature Importance from Logistic Regression coefficients

feature_importance = np.abs(lrm.coef_[0])
feature_names = features

# Sorting features by importance :

sorted_indices = np.argsort(feature_importance)[::-1]

print("Feature Importance(Top Churn Predictors) :")
for i in sorted_indices:
    print(f"{feature_names[i]}: {feature_importance[i]:.4f}")
```

# Ploting Feature Importance :

```python
sorted_importance = feature_importance[sorted_indices]
sorted_features = [feature_names[i] for i in sorted_indices]
```

```python
# Plotting Feature Importance :

plt.figure(figsize = (16,10))
sns.barplot(x = sorted_importance, y = sorted_features, palette = "tab10")
plt.xlabel("Feature Importance Score",fontsize = 15)
plt.ylabel("Features",fontsize = 15)
plt.title("Feature Importance for Churn Prediction")
plt.show()
```
